<p dir="rtl">נתייחס לגז כאל אוסף של N = 2000 כדורים קשיחים המפוזרים באקראי במישור (לשם פשטות, נתייחס בבעיה זו רק לשני מימדים).</p>
<p dir="rtl">לכל הכדורים אותה מסה m.</p>
<p dir="rtl">הנחת היסוד העיקרית של המודל שנבנה היא כי חלקיקי הגז הם כמו כדורים קשיחים המתנגשים אלסטית זה בזה.</p>
<p dir="rtl">נחלק את אוסף ה"כדורים" לשתי קבוצות.</p>
<p dir="rtl">עבור כל קבוצה נייצר מערך בגודל N × 2 המאחסן את רכיבי x ו-y של מהירות הכדורים.</p>
<p dir="rtl">נעניק מהירות התחלתית כלשהי לכדורים, כך שהמהירות הממוצעת היא 0.</p>
<p dir="rtl">למשל, אפשר לתת לכל הכדורים בקבוצה אחת מהירות של v₀ = 100 מטר לשנייה ולכל הכדורים בקבוצה השנייה מהירות של ‎−100‎ מטר לשנייה (בכיוון x).</p>
<p dir="rtl">נייצר מערך מאוחד של רכיבי x של המהירות ההתחלתית ונציג היסטוגרמה.</p>
<p dir="rtl">כעת נבנה לולאה של 100 איטרציות ובתוכה לולאה נוספת בה נגריל לכל זוג חלקיקים זווית פגיעה בתחום הערכים שבין 0 ל-π/2 (בעזרת np.random.default_rng().uniform).</p>
<p dir="rtl">נערוך התנגשות בין כל הכדורים מקבוצה 1 לכל הכדורים מקבוצה 2 עם זווית הפגיעה האקראית שהגרלנו ונעדכן את מערכי המהירויות לערכים של המהירות אחרי ההתנגשות (כדאי להעזר בתוצאה של התרגיל "התנגשות בין כדורים שווים”).</p>
<p dir="rtl">בסוף הלולאה "נערבב" את אחד ממערכי המהירות בעזרת np.random.default_rng().permutation</p>
<p dir="rtl">נאחד את מערכי המהירות ונציג היסטוגרמה של רכיבי x של המהירות.</p>
<p dir="rtl">יש להציג בגרף מספר היסטוגרמות של רכיבי המהירות בכיוון x עבור מספר איטרציות (נגיד עבור איטרציות 0, 1, 2, 3, 10, 20).</p>
<p dir="rtl">הסבירו באמצעות הגרפים מה מאפיין את התפלגות המהירויות כעבור מספיק זמן ומהו שיווי משקל תרמודינאמי.</p>
<p dir="rtl">שרטטו על ההיסטוגרמות גרף של התפלגות נורמלית  
f(vₓ) = 1/√(2πs) e^(-½(vₓ/s)²)  
עם s = √(kT/m),  
k = ‎1.38⋅10⁻²³‎ ג'אול לקלווין הוא קבוע בולצמן.  
עבור המסה ניקח m = ‎4.65⋅10⁻²⁶‎ קילוגרם.</p>
<p dir="rtl">ההתפלגות היא סביב ערך ממוצע ‎⟨vₓ⟩ = 0‎ וסטיית תקן ‎s = √⟨vₓ²⟩ - ⟨vₓ⟩²‎.</p>
<p dir="rtl">כיוון ש-‎⟨vₓ⟩ = 0‎ מתקבל ‎s² = ⟨vₓ²⟩‎ כלומר ‎kT = m⟨vₓ²⟩‎.</p>
<p dir="rtl">הסבירו מדוע ‎⟨vₓ² + v_y²⟩‎ אינו משתנה בעקבות ההתנגשויות.</p>
<p dir="rtl">כמו כן במצב שיווי המשקל ‎⟨vₓ²⟩ = ⟨v_y²⟩‎.</p>
<p dir="rtl">לכן נקבל כי ‎kT = m(v₀ₓ² + v₀_y²)‎.</p>
<p dir="rtl">כדי לשרטט את הפונקציה  
f(vₓ) = 1/√(2πs) e^(-½(vₓ/s)²)  
על גבי ההיסטוגרמה יש להשתמש בפרמטר density=True בהפקת ההיסטוגרמה.</p>
<p dir="rtl">כמו כן מומלץ להשתמש ב-bins=50.</p>
<p dir="rtl">הפעילו את הסימולציה פעמיים עבור שתי טמפרטורות שונות והציגו את שתי ההיסטוגרמות הסופיות המתקבלות בשני המקרים.</p>

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
rng = np.random.default_rng(42)

# Parameters
N = 2000  # Number of particles
m = 4.65e-26  # Mass of each particle [kg]
k = 1.38e-23  # Boltzmann constant [J/K]
v0_1 = 100  # Initial velocity for group 1 [m/s]
v0_2 = -100  # Initial velocity for group 2 [m/s]
iterations = 100  # Number of iterations
bins = 50  # Number of bins for histogram

# Initial velocities: two groups, each with N/2 particles
velocities1 = np.zeros((N//2, 2))
velocities2 = np.zeros((N//2, 2))
velocities1[:, 0] = v0_1  # Group 1: vx = +100 m/s, vy = 0
velocities2[:, 0] = v0_2  # Group 2: vx = -100 m/s, vy = 0

# Combine all vx for initial histogram
vx_all = np.concatenate([velocities1[:, 0], velocities2[:, 0]])

# Plot initial histogram
plt.figure(figsize=(8, 5))
plt.hist(vx_all, bins=bins, density=True, alpha=0.7, label='Iteration 0')
plt.xlabel('$v_x$ [m/s]')
plt.ylabel('Probability Density')
plt.title('Histogram of $v_x$ at Iteration 0')
plt.legend()
plt.show()
```

```python
# Function to perform elastic collision between two particles with random angle
def elastic_collision(v1, v2, theta):
    """
    v1, v2: velocity vectors of two particles (shape: (2,))
    theta: collision angle (radians)
    Returns: new velocity vectors after collision
    """
    # Rotate velocities to collision frame
    v1_rot = np.array([
        v1[0] * np.cos(theta) + v1[1] * np.sin(theta),
        -v1[0] * np.sin(theta) + v1[1] * np.cos(theta)
    ])
    v2_rot = np.array([
        v2[0] * np.cos(theta) + v2[1] * np.sin(theta),
        -v2[0] * np.sin(theta) + v2[1] * np.cos(theta)
    ])
    # Exchange x-components (elastic collision, equal mass)
    v1_rot[0], v2_rot[0] = v2_rot[0], v1_rot[0]
    # Rotate back to lab frame
    v1_new = np.array([
        v1_rot[0] * np.cos(-theta) + v1_rot[1] * np.sin(-theta),
        -v1_rot[0] * np.sin(-theta) + v1_rot[1] * np.cos(-theta)
    ])
    v2_new = np.array([
        v2_rot[0] * np.cos(-theta) + v2_rot[1] * np.sin(-theta),
        -v2_rot[0] * np.sin(-theta) + v2_rot[1] * np.cos(-theta)
    ])
    return v1_new, v2_new
```

```python
# Arrays to store histograms for selected iterations
histograms = {}
selected_iters = [0, 1, 2, 3, 10, 20]

# Main simulation loop
for it in range(iterations):
    # For each pair (i, i), collide velocities1[i] with velocities2[i]
    for i in range(N//2):
        theta = rng.uniform(0, np.pi/2)
        v1_new, v2_new = elastic_collision(velocities1[i], velocities2[i], theta)
        velocities1[i] = v1_new
        velocities2[i] = v2_new
    # At the end of each iteration, randomly permute velocities2
    velocities2 = rng.permutation(velocities2)
    # Store histograms for selected iterations
    if it in selected_iters:
        vx_all = np.concatenate([velocities1[:, 0], velocities2[:, 0]])
        histograms[it] = vx_all.copy()

# Plot histograms for selected iterations
plt.figure(figsize=(10, 6))
for it in [0, 1, 2, 3, 10, 20, 99]:
    if it in histograms:
        plt.hist(histograms[it], bins=bins, density=True, alpha=0.5, label=f'Iteration {it}')
plt.xlabel('$v_x$ [m/s]')
plt.ylabel('Probability Density')
plt.title('Histograms of $v_x$ at Selected Iterations')
plt.legend()
plt.show()
```

<p dir="rtl" align="right">
לאחר מספר איטרציות, ניתן לראות מהגרפים שההתפלגות של רכיב המהירות על ציר ה-x משתנה מהתפלגות דו-שיאית (שני פיקים בולטים בתחילה, כל אחד מייצג קבוצה עם מהירות התחלתית שונה) להתפלגות פעמונית סימטרית סביב אפס. ככל שמספר ההתנגשויות גדל, ההתפלגות מתקרבת יותר ויותר להתפלגות נורמלית (גאוסית), כפי שמוצג על ידי קו ההתפלגות התאורטית בגרפים.
</p>

<p dir="rtl" align="right">
<strong>שיווי משקל תרמודינאמי</strong> מתקבל כאשר ההתפלגות כבר אינה משתנה משמעותית בין איטרציה לאיטרציה, כלומר המערכת הגיעה למצב בו האנרגיה מתפזרת באופן אחיד בין כל החלקיקים, וההתפלגות של \( v_x \) היא נורמלית עם ממוצע אפס וסטיית תקן קבועה. זהו מצב בו אין העדפה לכיוון מסוים, והאנרגיה הקינטית הממוצעת של כל חלקיק שווה.
</p>

```python
# Calculate standard deviation s for the final distribution
vx_final = np.concatenate([velocities1[:, 0], velocities2[:, 0]])
s = np.std(vx_final)
mean_vx = np.mean(vx_final)

# Theoretical normal distribution for comparison
vx_range = np.linspace(-200, 200, 400)
normal_dist = (1 / (np.sqrt(2 * np.pi) * s)) * np.exp(-0.5 * ((vx_range - mean_vx) / s) ** 2)

# Plot final histogram and normal distribution
plt.figure(figsize=(8, 5))
plt.hist(vx_final, bins=bins, density=True, alpha=0.7, label='Final $v_x$ Histogram')
plt.plot(vx_range, normal_dist, 'r--', label='Normal Distribution')
plt.xlabel('$v_x$ [m/s]')
plt.ylabel('Probability Density')
plt.title('Final $v_x$ Histogram and Normal Distribution')
plt.legend()
plt.show()
```

```python
# Higher temp
v0_high = 200  
velocities1_high = np.zeros((N//2, 2))
velocities2_high = np.zeros((N//2, 2))
velocities1_high[:, 0] = v0_high
velocities2_high[:, 0] = -v0_high

for it in range(iterations):
    for i in range(N//2):
        theta = rng.uniform(0, np.pi/2)
        v1_new, v2_new = elastic_collision(velocities1_high[i], velocities2_high[i], theta)
        velocities1_high[i] = v1_new
        velocities2_high[i] = v2_new
    velocities2_high = rng.permutation(velocities2_high)

vx_final_high = np.concatenate([velocities1_high[:, 0], velocities2_high[:, 0]])
s_high = np.std(vx_final_high)
mean_vx_high = np.mean(vx_final_high)
normal_dist_high = (1 / (np.sqrt(2 * np.pi) * s_high)) * np.exp(-0.5 * ((vx_range - mean_vx_high) / s_high) ** 2)

# Plot comparison of final histograms for two temperatures
plt.figure(figsize=(8, 5))
plt.hist(vx_final, bins=bins, density=True, alpha=0.6, label='Regular temp')
plt.hist(vx_final_high, bins=bins, density=True, alpha=0.6, label='High temp')
plt.plot(vx_range, normal_dist, 'r--', label='Normal (Regular temp)')
plt.plot(vx_range, normal_dist_high, 'g--', label='Normal (High temp')
plt.xlabel('$v_x$ [m/s]')
plt.ylabel('Probability Density')
plt.title('Different temperatures')
plt.legend()
plt.show()
```
