# 🌟 Kinematics Formulas 🚀  

## 📏 Linear Kinematics  
1. **Displacement ($s$)**:  
   $$ s = ut + \frac{1}{2} a t^2 $$  
   where:  
   - $u$ = initial velocity  
   - $a$ = acceleration  
   - $t$ = time  

2. **Velocity ($v$)**:  
   $$ v = u + at $$

3. **Final Velocity Squared ($v^2$)**:  
   $$ v^2 = u^2 + 2as $$

4. **Average Velocity ($v_{\text{avg}}$)**:  
   $$ v_{\text{avg}} = \frac{u + v}{2} $$

---

## 🔄 Circular Kinematics  
1. **Angular Displacement ($\theta$)**:  
   $$ \theta = \omega t + \frac{1}{2} \alpha t^2 $$  
   where:  
   - $\omega$ = initial angular velocity  
   - $\alpha$ = angular acceleration  
   - $t$ = time  

2. **Angular Velocity ($\omega$)**:  
   $$ \omega = \omega_0 + \alpha t $$

3. **Centripetal Acceleration ($a_c$)**:  
   $$ a_c = \frac{v^2}{r} = \omega^2 r $$  
   where:  
   - $v$ = linear velocity  
   - $r$ = radius of the circular path  

4. **Tangential Velocity ($v_t$)**:  
   $$ v_t = \omega r $$

5. **Tangential Acceleration ($a_t$)**:  
   $$ a_t = \alpha r $$


```python
import numpy as np

import matplotlib.pyplot as plt

# Linear Kinematics
t = np.linspace(0, 10, 400)  # time array
u = 5  # initial velocity
a = 2  # acceleration

# Displacement
s = u * t + 0.5 * a * t**2

# Velocity
v = u + a * t

# Plotting Linear Kinematics
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, s, label='Displacement (s)')
plt.xlabel('Time (t)')
plt.ylabel('Displacement (s)')
plt.title('Linear Kinematics: Displacement vs Time')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocity (v)', color='orange')
plt.xlabel('Time (t)')
plt.ylabel('Velocity (v)')
plt.title('Linear Kinematics: Velocity vs Time')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, np.full_like(t, a), label='Acceleration (a)', color='green')
plt.xlabel('Time (t)')
plt.ylabel('Acceleration (a)')
plt.title('Linear Kinematics: Acceleration vs Time')
plt.legend()

plt.tight_layout()
plt.show()

# Circular Kinematics
omega_0 = 1  # initial angular velocity
alpha = 0.1  # angular acceleration
r = 3  # radius

# Angular Displacement
theta = omega_0 * t + 0.5 * alpha * t**2

# Angular Velocity
omega = omega_0 + alpha * t

# Tangential Velocity
v_t = omega * r

# Centripetal Acceleration
a_c = omega**2 * r

# Plotting Circular Kinematics
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(t, theta, label='Angular Displacement (θ)')
plt.xlabel('Time (t)')
plt.ylabel('Angular Displacement (θ)')
plt.title('Circular Kinematics: Angular Displacement vs Time')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t, omega, label='Angular Velocity (ω)', color='orange')
plt.xlabel('Time (t)')
plt.ylabel('Angular Velocity (ω)')
plt.title('Circular Kinematics: Angular Velocity vs Time')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t, v_t, label='Tangential Velocity (v_t)', color='green')
plt.xlabel('Time (t)')
plt.ylabel('Tangential Velocity (v_t)')
plt.title('Circular Kinematics: Tangential Velocity vs Time')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, a_c, label='Centripetal Acceleration (a_c)', color='red')
plt.xlabel('Time (t)')
plt.ylabel('Centripetal Acceleration (a_c)')
plt.title('Circular Kinematics: Centripetal Acceleration vs Time')
plt.legend()

plt.tight_layout()
plt.show()
```
